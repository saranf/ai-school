```python
def word_chain_game():
    used_words = []
    last_letter = ""

    while True:
        word = input("Enter a word (type 'exit' to quit): ").strip()

        if word == 'exit':
            print("Exiting the game.")
            break

        if not last_letter:
            used_words.append(word)
            last_letter = word[-1]
            print(f"Used word: {word}")
            continue

        if word in used_words:
            print(f"'{word}' has already been used. Game over!")
            break

        if word[0] != last_letter:
            print(f"The word doesn't start with '{last_letter}'. Game over!")
            break

        used_words.append(word)
        last_letter = word[-1]
        print(f"Used word: {word}")

word_chain_game()
```

```python
def save_words_to_file(word, filename="used_words.txt"):
    with open(filename, "a") as file:
        file.write(word + "\n")


def load_words_from_file(filename="used_words.txt"):
    try:
        with open(filename, "r") as file:
            return set(word.strip() for word in file.readlines())
    except FileNotFoundError:
        return set()


def word_chain_game():
    used_words = []
    last_letter = ""
    filename = "used_words.txt"

    while True:
        word = input("Enter a word (type 'exit' to quit): ").strip()

        if word == 'exit':
            print("Exiting the game.")
            break

        if len(used_words) >= 100:
            # 100개 이상이면 파일을 통해 중복 관리
            file_words = load_words_from_file(filename)
            if word in file_words:
                print(f"'{word}' has already been used (from file). Game over!")
                break
            else:
                save_words_to_file(word, filename)
        else:
            # 100개 이하일 때는 메모리에서 중복 관리
            if word in used_words:
                print(f"'{word}' has already been used (in memory). Game over!")
                break

        if not last_letter:
            used_words.append(word)
            last_letter = word[-1]
            print(f"Used word: {word}")
            continue

        if word[0] != last_letter:
            print(f"The word doesn't start with '{last_letter}'. Game over!")
            break

        used_words.append(word)
        last_letter = word[-1]
        print(f"Used word: {word}")

word_chain_game()
```

```python
fruits = ['사과','바나나','오렌지','포도','수박']
qtt = [3,3,3,3,3]
 
print("과일 가게에 오신 것을 환영합니다!")
while(True):
    search_flg = False
    if qtt == [0,0,0,0,0]:
        print("모든 과일이 다 팔렸습니다. 가게를 종료합니다.")
        break
    print("\n오늘 판매하는 과일: ", fruits)
    print("현재 재고: ",qtt)
    print("어떤 과일을 구매하시겠습니까?\n(종료하려면 '그만'을 입력하세요, 과일을 추가하시려면 '추가'를 입력하세요, 재고를 조정 하시려면 '조정'을 입력하세요.): ",end="")
    buy = input()
    print(buy)
    if buy == "그만":
        print("가게를 종료합니다.")
        break
    if buy == "추가":
        print("어떤 과일을 추가하시겠습니까? ",end="")
        add = input()
        print(add)
        print(add,"를 얼마만큼 추가하시겠습니까?")
        qtt_add = int(input())
        fruits.append(add)
        qtt.append(qtt_add)
        print(add,"을(를) ",qtt_add,"개 추가했습니다.")
        continue
    if buy == "조정":
        print("어떤 조정을 하시겠습니까? 이름을 수정하시겠으면 '이름', 개수를 수정하시겠으면 '개수'를 입력하세요.",end="")
        cmd = input()
        print(cmd)
        if cmd == "이름":
            print("수정하실 과일 이름을 입력해주세요.",end="")
            name = input()
            print(name)
            for i in fruits:
                if name == i:
                    search_flg = True
                    print("어떤 이름으로 수정하시겠습니까?",end="")
                    new_name = input()
                    print(new_name)
                    fruits[fruits.index(i)] = new_name
                    print(f"{i}를 {new_name}으로 수정 하였습니다.")
            if search_flg == False:
                print("해당 과일은 존재하지 않습니다.")
        elif cmd == "개수":
            print("수정하실 과일 이름을 입력해주세요.",end="")
            name = input()
            print(name)
            for i in fruits:
                if name == i:
                    search_flg = True
                    print("몇개로 수정하시겠습니까?",end="")
                    new_qtt = int(input())
                    print(new_qtt)
                    qtt[fruits.index(i)] = new_qtt
                    print(f"{i}를 {new_qtt}개로 수정 하였습니다.")
            if search_flg == False:
                print("해당 과일은 존재하지 않습니다.")
        else:
            print("잘못입력하셨습니다.")
            continue
    for i in fruits:
        if buy == i:
            if qtt[fruits.index(i)] == 0:
                print("해당 과일은 매진되었습니다.")
                search_flg = True
            else:
                qtt[fruits.index(i)] -= 1
                print(f"{i}을(를) 구매하셨습니다! 남은 수박의 재고: {qtt[fruits.index(i)]}")
                search_flg = True
    if search_flg == False:
        print("해당 과일은 존재하지 않습니다.")
   
```

```python
import re
 
phone_num_pattern = re.compile(r'^\d{3}-\d{4}-\d{4}$')
 
tele_book = {}
 
while True:
    menu = input("메뉴 선택(1:추가, 2:검색, 3:삭제, 4:종료)? ")
    if menu == "1":
        name = input("* 친구 이름 >> ")
        phone_num = input("* 전화번호 >> ")
        if phone_num_pattern.match(phone_num):
            pass
        else:
            print("잘못된 전화번호 형식")
            continue
        if name in tele_book:
            yn = input("* 이미 입력된 이름, 수정 여부? y/n : ")
            if yn == "y":
                pass
            else:
                continue
        tele_book[name] = phone_num
    elif menu == "2":
        print(f"검색 가능 리스트 : {tele_book.keys()}")
        if len(tele_book.keys()) <= 0:
            print("검색할 이름 없음")
            continue
        name = input("* 친구 이름 >> ")
        print(f"{name}의 전화번호 : {tele_book[name]}")
    elif menu == "3":
        name = input("* 친구 이름 >> ")
        if name not in tele_book:
            print("목록에 없는 이름")
            continue
        else:
            tele_book.pop(name)
    elif menu == "4":
        print("프로그램을 종료합니다.")
        break
    else:
        print("메뉴 선택 오류입니다. 다시 선택하세요.")
 
```
